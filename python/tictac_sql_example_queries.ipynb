```python
import mysql.connector as connector
from mysql.connector import Error
import pandas as pd

def execute_query(connection, query):
    try:
        cursor = connection.cursor()
        cursor.execute(query)
        results = cursor.fetchall()
        column_names = [i[0] for i in cursor.description]
        df = pd.DataFrame(results, columns=column_names)
        return df
    except Error as e:
        print(f"Error: {e}")
        return None

def main():
    try:
        # Establish the connection
        connection = connector.connect(
            host="localhost",         # Replace with your actual host if different
            user="root",     # Replace with your actual username
            password="Je@#1978", # Replace with your actual password
            database="tictac_db"          # Replace with your actual database name
        )

        if connection.is_connected():
            # Fetch and print some server information
            db_info = connection.get_server_info()
            print("Connected to MySQL Server version ", db_info)
            
            # Verify the connection
            cursor = connection.cursor()
            cursor.execute("SELECT DATABASE();")
            record = cursor.fetchone()
            print("You're connected to database: ", record)
            
            # Optional: Show tables in the connected database
            cursor.execute("SHOW TABLES;")
            tables = cursor.fetchall()
            print("Tables in the database:", tables)
            
            # Queries
            queries = {
                "Study details for Calcitonin receptor in Breast cancer": """
                SELECT * FROM aact_studies WHERE nct_id IN
                (SELECT DISTINCT t.nct_id
                    FROM tictac_genes_diseases_associations AS t
                    JOIN aact_studies AS s ON t.nct_id = s.nct_id
                    WHERE t.tcrdTargetName LIKE '%Calcitonin receptor%'
                    AND t.disease_name = 'Breast cancer');
                """,
                "Highest mean rank score association": """
                SELECT *
                FROM tictac_genes_diseases_associations
                ORDER BY mean_rank_score DESC
                LIMIT 1;
                """,
                "Least mean rank score association": """
                SELECT *
                FROM tictac_genes_diseases_associations
                ORDER BY mean_rank_score ASC
                LIMIT 1;
                """,
                "Least mean rank score association (non-null)": """
                SELECT *
                FROM tictac_genes_diseases_associations
                WHERE mean_rank_score IS NOT NULL
                ORDER BY mean_rank_score ASC
                LIMIT 1;
                """,
                "Top 25 highest mean rank score associations": """
                SELECT *
                FROM tictac_genes_diseases_associations
                WHERE mean_rank_score IS NOT NULL
                ORDER BY mean_rank_score DESC
                LIMIT 25;
                """,
                "Heart disease associations with highest mean rank score of 100": """
                SELECT *
                FROM tictac_genes_diseases_associations
                WHERE disease_name LIKE '%heart%'
                  AND mean_rank_score = 100
                ORDER BY mean_rank_score DESC
                LIMIT 10;
                """
            }

            # Execute and display results for each query
            for description, query in queries.items():
                print(f"\nExecuting query: {description}")
                df = execute_query(connection, query)
                if df is not None:
                    print(df)
                else:
                    print("No results returned or error in query execution")

    except Error as e:
        print("Error while connecting to MySQL", e)

    finally:
        if connection.is_connected():
            cursor.close()
            connection.close()
            print("MySQL connection is closed")

if __name__ == "__main__":
    main()
```
